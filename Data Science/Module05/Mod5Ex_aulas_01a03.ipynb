 7) seu código aqui
 
 A Coluna LOCNASC tem apenas 4 possibilidades, não pode ser feita com um valor mais provável,
pois cada número contém um significado de uma possível região,  o mesmo se aplica a GRAVIDEZ, não pode ser mensurado um valor próximo.
````python
# 8) seu código aqui
sinasc_reduz.loc[sinasc_reduz['APGAR5'] <= 3.0, 'Apgar_cat' ] = 'asfixia severa'
sinasc_reduz.loc[(sinasc_reduz['APGAR5'] > 4.0) & (sinasc_reduz['APGAR5'] <= 5.0), 'Apgar_cat' ] = 'asfixia moderada'
sinasc_reduz.loc[(sinasc_reduz['APGAR5'] > 5.0) & (sinasc_reduz['APGAR5'] <= 7.0), 'Apgar_cat' ] = 'aasfixia leve'
sinasc_reduz.loc[sinasc_reduz['APGAR5'] >= 8.0, 'Apgar_cat' ] = 'normal'

sinasc_reduz
````
