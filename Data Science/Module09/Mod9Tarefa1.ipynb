# Tarefa 1

### 1. Idade media das mães e dos pais por município (coluna munResNome)


```python
sinasc = sinasc_raw[['munResNome', 'IDADEMAE', 'IDADEPAI']]
```

```python
#round(sinasc.groupby('munResNome')['IDADEMAE', 'IDADEPAI'].mean(), 2)
sinasc.groupby('munResNome').aggregate({'IDADEMAE' : 'mean', 'IDADEPAI': 'mean'})
```

### 2. Peso médio dos bebes por sexo que nasceram no dia do seu aniversário por faixas de escolaridade mae
Ex: Você, aluna(o), nasceu no dia 10/01, então você precisa filtrar o conjunto de dados nessa data e calcular o peso médio dos bebês de cada sexo por faixa de escolaridade da mãe.

```python
sinasc_data = sinasc_raw[['DTNASC', 'PESO', 'ESCMAE','SEXO']]
sinasc_data.head()
```

```python
filtro = sinasc_data[sinasc_data['DTNASC'] == '2019-06-01']
filtro.groupby(['ESCMAE','SEXO'])['PESO'].mean()
```

### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
sinasc_idade = sinasc_raw[['munResNome', 'IDADEMAE', 'IDADEPAI']]
sinasc_idade.head()
```

```python
sinasc_idade.groupby('munResNome')['IDADEMAE'].count().sort_values()
```

```python
filtro_id = sinasc_idade[sinasc_idade['munResNome'] == 'Castanheiras']
filtro_id.groupby('munResNome').aggregate({'IDADEMAE' : [('mínimo', min),('média', 'mean'),('máxima', max)],
                                           'IDADEPAI' : [('mínimo', min),('média', 'mean'),('máxima', max)]})

```

#### Castanheiras tiveram tanto com a idade minima e máxima das mães e dos pais, muito parecidas, a média vario em 3 anos.

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



```python
sinasc_raw['DTNASC'] = pd.to_datetime(sinasc_raw['DTNASC'])
sinasc_raw['month']= sinasc_raw['DTNASC'].dt.month
sinasc_raw['month']
```

```python
filtro_mar = sinasc_raw[sinasc_raw['month'] == 3]
filtro_mar.groupby('DTNASC').aggregate({'QTDFILVIVO' : [('média','mean'),('máxima',max),('mínima',min)]})
```

```python
filtro_mar.groupby('DTNASC').aggregate({'IDADEPAI' : [('média','mean'),('máxima',max),('mínima',min)]})
```

#### Existe uma grande amplitude entre a idade máxima e mínima no mês de março, apenas o último dia que teve uma amplitude pequena (max = 34 & min = 19)
