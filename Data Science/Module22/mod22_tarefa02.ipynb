#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
yf_fechamento = yf['Adj Close']
yf_fechamento.head()
````

````python
yf_fechamento90 = yf_fechamento.rolling(90).mean()
yf_fechamento90_vertical = yf_fechamento90.stack().reset_index()
yf_fechamento90_vertical.columns = ['data','ativo','valor']
````

````python
yf_fechamento_std = yf_fechamento.rolling(90).std()
yf_fechamento_stdvertical = yf_fechamento_std.stack().reset_index()
yf_fechamento_stdvertical.columns = ['data','ativo','valor']
````

````python
fig = px.line(yf_fechamento90_vertical, x="data", y="valor", color='ativo', title='Média Movel 90')
fig.show()
````

````python
fig = px.line(yf_fechamento_stdvertical, x="data", y="valor", color='ativo', title = 'Desvio Padrão')
fig.show()
````

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
lista_indicadores = ['Close', 'High', 'Low', 'Open']
ativo = ['ITSA4.SA']*len(lista_indicadores)
````

````python
colunas = list(zip(lista_indicadores, ativo))
````

````python
itausa = yf.loc['01/01/2022':, colunas]
itausa.columns = lista_indicadores
````

````python
itausa.head()
````

````python
graph = {
    'x' : itausa.index,
    'open'  : itausa.Open,
    'close' : itausa.Close,
    'high'  : itausa.High,
    'low'   : itausa.Low,
    'type'  : 'candlestick',
    'name'  : 'ITSA4.SA',
    'showlegend':True
}
````

````python
from plotly import graph_objects
from plotly.graph_objects import Layout

layout = Layout(
    paper_bgcolor = 'rgba(0,0,0,0)',
    plot_bgcolor  = 'rgba(0,0,0,0)'
)

grafico = graph_objects.Figure(
    data = [graph],
    layout_title = 'Itausa',
    layout = layout
)

grafico.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')
grafico.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')

grafico
````
