#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
#'tipo_renda', 'educacao, estado_civil', 'tipo_residencia'
base = base.drop(columns=['estado_civil'])
base = base.drop(columns=['tipo_residencia'])

base
````

````python
base_encoded = pd.get_dummies(base, columns = ['tipo_renda', 'educacao'], drop_first=True)
base_encoded
````

````python
base_encoded['sexo'].unique()
````

````python
base_encoded.sexo = base_encoded.sexo.map({'M': 1, 'F': 0})
base_encoded['sexo'].unique()
````

````python
base_encoded['posse_de_veiculo'].unique()
````

````python
base_encoded.posse_de_veiculo = base_encoded.posse_de_veiculo.map({'Y': 1, 'N': 0})
base_encoded['posse_de_veiculo'].unique()
````

````python
base_encoded['posse_de_imovel'].unique()
````

````python
base_encoded.posse_de_imovel = base_encoded.posse_de_imovel.map({'Y': 1, 'N': 0})
base_encoded['posse_de_imovel'].unique()
````

````python
base_encoded.mau = base_encoded.mau.astype(int)
base_encoded
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
freq = pd.crosstab(base_encoded['posse_de_veiculo'], base_encoded['possui_email'])
freq.sum(axis=1)


# Possuir ou não um email, não reflete na situação da nossa variavél resposta,
# já possuir um veiculo pode trazer mais resultados visando que ele demandou um dinheiro na compra, e o pagamento do carro
# pode ser um motivo para ele ter se tornado um mau pagador.
````
